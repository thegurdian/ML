```python
import pandas as pd

data = {
    "experience_years": [1, 2, 3, 4, 5, 6, 7, 8],
    "salary": [15000, 20000, 30000, 50000, 80000, 120000, 200000, 500000]
}

df = pd.DataFrame(data)
print(df)
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler

# Separate features
X = df[["experience_years", "salary"]]



# MinMaxScaler
minmax_scaler = MinMaxScaler()
X_minmax = minmax_scaler.fit_transform(X)




df_minmax = pd.DataFrame(X_minmax, columns=X.columns)


print("\nMinMax Scaled:\n", df_minmax)
```
