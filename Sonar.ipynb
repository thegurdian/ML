```python
X=df.drop(columns="R",axis=1)
```

```python
Y=df["R"]
```

Now that we have separated the features (`X`) and the target variable (`Y`), we need to split the data into training and testing sets. This allows us to train our model on one part of the data and evaluate its performance on unseen data. We'll use `train_test_split` from `sklearn.model_selection` for this.

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

# Re-split X and Y to ensure Y_train and Y_test are fresh string Series
# Y is assumed to be `df["R"]` as defined in a previous cell.
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)

encoder = LabelEncoder()
Y_train = encoder.fit_transform(Y_train)  # Learn mapping
Y_test = encoder.transform(Y_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model=Sequential([
    Dense(64,activation="relu",input_shape=(X_train.shape[1],)),
    Dense(32,activation="relu"),
    Dense(8, activation="relu"),
    Dense(1, activation="sigmoid")

])
```

```python
model.compile(
    optimizer="adam",
    loss="binary_crossentropy",
    metrics=["accuracy"]
)
```

```python
history = model.fit(
    X_train,
    Y_train, # Use the encoded labels
    validation_split=0.2,
    epochs=20,
    batch_size=32,

)
```

```python
model.save("sonar_ann.h5")
```
