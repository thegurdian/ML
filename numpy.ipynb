```python
import numpy as np

A = np.array([[1,2],
              [3,4]])

B = np.array([[5,6],
              [7,8]])

print(np.dot(A, B))
print(A @ B)
```

```python
A = np.array([[1,2],
              [3,4]])

A_inv = np.linalg.inv(A)
print(A_inv)
```

```python
A = np.array([[2,0],
              [0,3]])

values, vectors = np.linalg.eig(A)
print(values)
print(vectors)
```
