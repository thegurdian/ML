ReLU activation function


```python
import numpy as np



def relu(x):
  return np.maximum(0,x)

relu(np.array([-2,10,4,6]))
```

Leaky ReLU Function

```python
def leaky_relu(x,alpha=0.01):
  return np.where(x>0,x,alpha*x)

leaky_relu(np.array([-19,2,3,4]))
```

# ELU

```python
def ELU(x,alpha=0.01):
  return np.where(x>0,x,alpha *(np.exp(x)-1))
ELU(np.array([-10,1,2,3]))
```

# Sigmoid

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = np.array([-3, -1, 0, 1, 3])

print(sigmoid(x))
```

# Tanh Function

```python
def tanh(x):
    return np.tanh(x)

x = np.array([-3, -1, 0, 1, 3])

print(tanh(x))
```

# Swish

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def swish(x):
    return x * sigmoid(x)

x = np.array([-3, -1, 0, 2])

print(swish(x))
```
