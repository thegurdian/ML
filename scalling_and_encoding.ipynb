```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
```

```python
data = {
    'Name':['Ali','Sara','Rahim','Nila','Karim'],
    'Gender':['Male','Female','Male','Female','Male'],
    'City':['Dhaka','Chattogram','Dhaka','Sylhet','Rajshahi'],
    'Age':[22,28,35,30,40],
    'Salary':[25000,40000,55000,45000,70000]
}
```

```python
df=pd.DataFrame(data)
```

label encoding

```python
le=LabelEncoder()
df['gender']=le.fit_transform(df['Gender'])
```

one hot encoding


```python
oh=OneHotEncoder()
encoded=oh.fit_transform(df[['City']]).toarray()
encoded_df = pd.DataFrame(
    encoded,
    columns=oh.get_feature_names_out(['City'])
)

df = pd.concat([df, encoded_df], axis=1).drop('City', axis=1)

print(df)
```
