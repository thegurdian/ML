```python
from sklearn.preprocessing import OrdinalEncoder
import pandas as pd

df = pd.DataFrame({
    'experience': ['Junior', 'Mid', 'Senior', 'Junior']
})

encoder = OrdinalEncoder()
df['experience_encoded'] = encoder.fit_transform(df[['experience']])

print(df)
```
